```python
import json
```

```python
file = open('/home/jupyter/robo-romeo/data/annotations/captions_train2017.json', 'r')
dataset = json.load(file)
```

```python
dict_ids = {value['id']:value['file_name']  for value in dataset['images']}
```

```python
file = []
l_captions_sorted_image=sorted(dataset['annotations'],key=lambda x:x['image_id'])
for caption in l_captions_sorted_image:
    file.append(dict_ids[caption['image_id']] + ' ' + caption['caption'])
```

```python
type(file), len(file)
```

```python
lines = ('uniqueunique').join(file)
file = open('/home/jupyter/robo-romeo/data/captions.txt', 'w')
file.write(lines)
file.close()
```

```python
imgid_cap_dict = {}
for caps in dataset['annotations']:
    if str(caps['image_id']) not in imgid_cap_dict.keys():
        imgid_cap_dict[str(caps['image_id'])] = list()
    imgid_cap_dict[str(caps['image_id'])].append(caps['caption'])
```

```python
imgid_file_dict = {}
for imgs in dataset['images']:
    imgid_file_dict[str(imgs['id'])] = imgs['file_name']
```

```python
file_cap_dict = {}
for ids, caps in imgid_cap_dict.items():
    file_cap_dict[str(imgid_file_dict[ids])] = caps
```
